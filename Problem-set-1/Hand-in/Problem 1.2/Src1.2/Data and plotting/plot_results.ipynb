```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Plot training data

data = pd.read_csv('trainingData.csv')

data = data[data['Epoch'] < 40]

plt.figure(figsize=(8, 4), dpi=150)
plt.plot(data['Epoch'], data['TrainingAccuracy'], label='Training Accuracy')
plt.plot(data['Epoch'], data['ValidationAccuracy'], label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Training and Validation Accuracy of Perceptron Classifier')
plt.legend()
plt.grid(True)
plt.show()

plt.savefig("trainingProgress.png", dpi=300, bbox_inches='tight')
```
